# Series

# Creating a pandas 
there are two ways of creating series 
1. using Python list
2. using Dictionary

```python

import pandas as pd
```

Python List

```python
# we will use Series function to create a pandas series
# make sure that the letter 'S' in Series is capital
df = pd.Series([200,600,800,400,1000])
print(df)
```

*Labels*<br>
Labels are called as index in pandas

```python
df = pd.Series([200,300,400,500,600])
# accessing the first element of the series
print(df[0])
```

```python
# while creating an array we can specify our own index(labels) for the series
# for that we have to parameter use  index 
df = pd.Series([200,300,400,500,600],index=['a','b','c','d','e'])
print(df)
```

```python
# Accessing the element using labels(index)
print(df['a'])
```

```python
# when specifying the label(index) we use .index method to assign the labels to the series
df = pd.Series([200,300,400,500,600])
df.index = ['a','b','c','d','e']
print(df)
```

```python
# while we are specifying the index for the values in the series we can also change the index after creating the series
df = pd.Series([200,300,400,500,600])
df.index = ['x','y','z','p','q']
print(df)
```

```python
# we can specify the data type of the values in the series using dtype parameter
df = pd.Series([200,300,400,500,600],index=['a','b','c','d','e'],dtype = float)
print(df)
```
