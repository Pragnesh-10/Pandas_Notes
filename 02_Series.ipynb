# Series

# Creating a pandas Series
there are two ways of creating series 
1. using Python list
2. using Dictionary

```python

import pandas as pd
```

Python List

```python
# we will use Series function to create a pandas series
# make sure that the letter 'S' in Series is capital
df = pd.Series([200,600,800,400,1000])
print(df)
```

*Labels*<br>
Labels are called as index in pandas

```python
df = pd.Series([200,300,400,500,600])
# accessing the first element of the series
print(df[0])
```

```python
# while creating an array we can specify our own index(labels) for the series
# for that we have to parameter use  index 
df = pd.Series([200,300,400,500,600],index=['a','b','c','d','e'])
print(df)
```

```python
# Accessing the element using labels(index)
print(df['a'])
```

```python
# when specifying the label(index) we use .index method to assign the labels to the series
df = pd.Series([200,300,400,500,600])
df.index = ['a','b','c','d','e']
print(df)
```

```python
# while we are specifying the index for the values in the series we can also change the index after creating the series
df = pd.Series([200,300,400,500,600])
df.index = ['x','y','z','p','q']
print(df)
```

```python
# we can specify the data type of the values in the series using dtype parameter
df = pd.Series([200,300,400,500,600],index=['a','b','c','d','e'],dtype = float)
print(df)
```

```python
# we can also specify the data type of the values in the series using astype() method
df = pd.Series([200,300,400,500,600],index=['a','b','c','d','e'])
df = df.astype(int)
print(df)
```

```python
# we can set the name of the a series using name parameter
df = pd.Series([200,300,400,500,600], name="Land Size")
print(df)
```

```python
# to get the name of the series we will use variablename.name
print(df.name)
```

```python
# we can change the name of the series using name attribute
df.name = "Area of Land"
print(df)
```

# Create Series From a Python Dictionary

```python
marks = {'Maths':90,'Science':80,'English':85}
x = pd.Series(marks)
print(x)
# here first we have a dictionary and later we have converted that into pandas Series
# here the keys of the dictionary are the labels(index) of the series
```

```python
# we can access the elements of the series by using the labels(index) of the series
marks = {'Maths':90,'Science':80,'English':85}
x = pd.Series(marks,index=['Maths','science'])
print(x)
```
