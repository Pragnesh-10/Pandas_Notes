```python
import pandas as pd
import numpy as np
```

```python

df = pd.read_csv('student_performance.csv')
print(df.head())
```

```python
# here we dropping final percentage column for showing how to add new columns
df.drop(columns=['Final_Percentage'], inplace=True)
print(df.head())
```

# add new column

```python
df['Average']= (df['Math_Score'] + df['Science_Score'] + df['English_Score']) / 3
print(df.head())
# if there are any NaN(none) values in any of the row then average column will have NaN value for that row
```

```python
# here we are calculation percentage 
df['Percentage'] = (df['Average']*300)/100
print(df.head())
```

```python
# we can round the percentage column values to 2 decimal places using round() method
df['Average'] = df['Average'].round(2)
print(df.head())
```

# adding new coulumn by using np.select

```python
cond = [
    (df['Percentage'] >= 90),
    (df['Percentage'] >= 75) & (df['Percentage'] < 90),
    (df['Percentage'] >= 50) & (df['Percentage'] < 75),
    (df['Percentage'] < 50) & (df['Percentage'] >= 35),
    (df['Percentage'] < 35) & (df['Percentage'] == 0),
    (df['Percentage'].isnull())
]
choice = ['A+', 'A', 'B', 'C', 'F', 'Not Available']

```

```python
# some of the values in percentage column are NaN so we have to handle that while assigning grades for that we have to use default parameter
df['Grades'] = np.select(cond, choice, default='Not Available')
print(df.head())
# so if you observe the grades column has Not Available for the rows which have NaN values in percentage column because we have assigned default value as Not Available for the rows which have NaN values in percentage column
# if we wont use default parameter then it will raise an error because there are some NaN values in percentage column so we have to handle that while assigning grades for that we have to use default parameter
```
