# Drop Row

```python
import pandas as pd
```

```python
df = pd.read_csv('student_performance.csv')
print(df.head())
```

```python
# we drop rows in two ways one is by using axis parameter and the other is by using column parameter
df.drop('Final_Percentage',axis = 1, inplace = True)
print(df.head())
# so if you observe the output here we have dropped final percentage column
```

```python
# here is the using dropping columns using columns parameter
df.drop(columns = ['Gender'], inplace = True)
print(df.head())
```

```python
# we can also drop multiple columns at once
df.drop(columns = ['Math_Score', 'Science_Score'], inplace = True)
print(df.head())
```

```python
# we can also drop collumns by using index but it is not recommended because if the index changes then the code will break
df.drop(df.columns[[1]], axis = 1, inplace = True)
print(df.head())
# so if you observe the output here we have dropped the second column
```

```python
# we can also use negative indexing to drop the last column
df.drop(df.columns[-1], axis = 1, inplace = True)
print(df.head())
# so if you observe the output here we have dropped the last column 
```

# Drop Row

```python
df = pd.read_csv('student_performance.csv')
print(df.head())
```

```python
# we can drop rows by using index parameter but it is not recommended because if the index changes then the code will break 
df.drop(index = [0, 1], inplace = True)
print(df.head())
```

```python
# here we are setting index to student id column
df.set_index('Student_ID', inplace = True)
```

```python
# her we are dropping rows by using student id as index
df.drop(index=['S0003'], inplace=True)
print(df.head())
```

```python
# here we are dropping multiple rows at once by using student id as index
df.drop(['S0004','S0005'], axis=0, inplace=True)
print(df.head())
```
