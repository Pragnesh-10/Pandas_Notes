# concat()

concatenation:<br>
The operation of joining two or more character strings, values, or data sets end-to-end to create a single, longer string or sequence

```python
df1 = pd.DataFrame({'id': ['A', 'B', 'C', 'D'],
                    'age': [30, 23, 25, 22]})
df2 = pd.DataFrame({'id': ['E', 'F', 'G', 'F'],
                    'age': [40, 21, 19, 24]})
```

```python
# tho concatenate the two dataframes we can use concat() method of pandas library 
pd.concat([df1, df2])
```

```python
# if you see the aboves example output you will see that the index is repeated because we are concatenating along rows and by default the index is not reseted so to reset the index we can use ignore_index parameter and set it to True
pd.concat([df1, df2], ignore_index=True)
```

```python
# we can specify the axis parameter to concatenate along rows or columns
# to concatenate along rows we have to use axis = 0 
# by default the aixs is 0
pd.concat([df1, df2], axis=0, ignore_index=True)
```

```python
# to concatenate along columns we have to use axis = 1
pd.concat([df1, df2], axis=1)
```

# merge()

merge means <br>
to combine or join

```python
df1 = pd.DataFrame({
    'key': ['K0', 'K1', 'K2', 'K3'],
    'Name': ['Jai', 'Princi', 'Gaurav', 'Anuj'],
    'Age': [27, 24, 22, 32],
})

df2 = pd.DataFrame({
    'key': ['K0', 'K1', 'K2', 'K3'],
    'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
    'Qualification': ['Btech', 'B.A', 'Bcom', 'B.hons'],
})
```

```python
# here we are merging the two dataframes on the common column 'key' and by default it will perform an inner join which means it will only keep the rows that have matching keys in both dataframes
# here if we observe the two dataframes we have two one common column which is 'key'
# to merge by using the common column we can use on parameter and the name of the commom column
pd.merge(df1, df2, on='key')
```

```python
pd.merge(df1, df2, on='key', how='inner')  # inner join
```

```python

pd.merge(df1, df2,on = 'key', how='outer')  # outer join
            
```
