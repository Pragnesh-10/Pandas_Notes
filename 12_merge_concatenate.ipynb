# concat()

concatenation:<br>
The operation of joining two or more character strings, values, or data sets end-to-end to create a single, longer string or sequence

```python
df1 = pd.DataFrame({'id': ['A', 'B', 'C', 'D'],
                    'age': [30, 23, 25, 22]})
df2 = pd.DataFrame({'id': ['E', 'F', 'G', 'F'],
                    'age': [40, 21, 19, 24]})
```

```python
# tho concatenate the two dataframes we can use concat() method of pandas library 
pd.concat([df1, df2])
```

```python
# if you see the aboves example output you will see that the index is repeated because we are concatenating along rows and by default the index is not reseted so to reset the index we can use ignore_index parameter and set it to True
pd.concat([df1, df2], ignore_index=True)
```

```python
# we can specify the axis parameter to concatenate along rows or columns
# to concatenate along rows we have to use axis = 0 
# by default the aixs is 0
pd.concat([df1, df2], axis=0, ignore_index=True)
```

```python
# to concatenate along columns we have to use axis = 1
pd.concat([df1, df2], axis=1)
```

# merge()

merge means <br>
to combine or join

```python
import pandas as pd

customers = pd.DataFrame({
    "customer_id": [1, 2, 3],
    "name": ["A", "B", "C"]
})

orders = pd.DataFrame({
    "customer_id": [1, 1, 2],
    "order_amount": [250, 300, 150]
})
```

```python
# here we are merging the two dataframes on the common column 'customer_id' and by default it will perform an inner join which means it will only keep the rows that have matching keys in both dataframes
# here if we observe the two dataframes we have one common column which is 'customer_id'
# to merge by using the common column we can use on parameter and the name of the common column
pd.merge(customers, orders, on='customer_id')
```

```python
# This is the default merge method in pandas. It acts like an intersection in set theory, keeping only the records that have a match in both the left and right datasets. 

pd.merge(customers, orders, on='customer_id', how='inner')  # inner join
```

```python
# This is a full outer join. It combines the data, keeping every single row from both DataFrames. If a row from one DataFrame doesn't have a matching key in the other, it fills in NaN
pd.merge(customers,orders,on = 'customer_id', how='outer')              
```

```python
# when we are combining two datasets, to keep all rows from the left table and only matching rows from the right table
# if there is no match in the right table then the result will have NaN values for the columns of the right table
pd.merge(customers, orders, on='customer_id', how='left')
```

```python
# when we are combining two datasets, to keep all rows from the right table and only matching rows from the left table
# if there is no match in the left table then the result will have NaN values for the columns of the left table
pd.merge(customers, orders, on="customer_id", how="right")
# A RIGHT merge shows NaN only when the right table has keys missing in the left table
```
