```python
import pandas as pd
```

```python
# read_csv() is used to read csv file and create a DataFrame from it
# we can also use read_excel() to read excel files and create a DataFrame from it
# we can also use read_json() to read json files and create a DataFrame from it
df = pd.read_csv('student_performance.csv')
print(df)
```

```python
# to print the entire DataFrame we can use to_string() method
print(df.to_string())
```

```python
# to print the first 5 rows we can use head() method
print(df.head())
```

```python
# by default it print the first 5 rows if we didnt specify the number of rows to print
# we can also specift the number of rows we want to print by number in head() method
print(df.head(7))
```

```python
# to print the last 5 rows we use tail() method by default it prints last 5 rows
print(df.tail())
```

```python
# we can also the specify the number of rows we want to print in tail() method
print(df.tail(3)) 
```
