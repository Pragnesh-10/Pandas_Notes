# value_counts()
Return a Series containing counts of unique values.

```python
import pandas as pd

df = pd.DataFrame({
    "gender": ["Male","Female","Male","Female","Male","Male"],
    "payment_mode": ["UPI","COD","UPI","Card","UPI","COD"],
    "age": [23, 25, 23, 30, None, 25],
    "city": ["Delhi","Delhi","Mumbai","Delhi","Mumbai","Chennai"],
    "status": ["active","inactive","active","active","active","inactive"]
})
print(df)
```

```python
# here it gives the count of unique values in the gender column
df["gender"].value_counts()
```

```python
# here it gives the count of unique values in the payment_mode column and also gives the percentage of each unique value in the payment_mode column
df["payment_mode"].value_counts(normalize=True)
# normalise parameter default value is False
```

```python
# we can also use multiple columns in value_counts() 
df.value_counts(["city", "payment_mode"])
```

```python
# here it gives also gives nan values if use use this parameter dropna=False in value_counts() method
df["age"].value_counts(dropna=False)
# dropna default vale is True
```
