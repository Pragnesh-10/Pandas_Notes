# isin()
it used to check whether the values are there in specific column or not

```python
import pandas as pd
```

```python
df = pd.read_csv('student_performance.csv')
print(df)
```

```python
# here we are checking if the math score is 80 or not for all the rows in the math score column and it will return a boolean value for each row
df['Math_Score'].isin([80])
```

```python
# we can use isin() method for multiple values
df['Math_Score'].isin([80, 90])
```

```python
# we can use the boolean masking to filter the rows where the math score is 80 to get the actual rows where the math score is 80
df[df['Math_Score'] == 80]
```

```python
# we can also use loc[] method to filter the rows where the math score is 80
df.loc[df['Math_Score'] == 80]
```

# isin() multiple filtering

```python
# 
f1 = df['Math_Score'] > 80
f2 = df['Science_Score'] > 80
df[f1&f2]
```
