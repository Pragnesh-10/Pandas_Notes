# loc and iloc
loc is used for slecting rows or columns based on labels(index) of them<br>
iloc is used selection rows and columns based on integer position of them

```python
df = pd.DataFrame({
    'Name' : ['Ninja_1','Ninja_2','Ninja_3','Ninja_4'],
    'Department' : ['Marketing','Sales','Marketing','Engineering'],
    'Salary' : [50000,60000,55000,70000]

})
print(df)                                             
```

the index of Name is 0, index of Department is 1, index of Salary is 2

```python
# Selecting A Single 
df.loc[2]
```

```python
# Selecting Multiple Rows
df.loc[[1, 2, 3]]
# As you can see, the rows with labels '1','2','3' are selected. Here it is indicating towards the second, third and fourth rows in DataFrame since Python uses 0-based indexing technique.
```

```python
# Selecting Multiple Columns
df.loc[:, ['Name', 'Salary']]
```

```python
# Selecting A Subset Of Rows And Columns
df.loc[1:2, ['Name', 'Salary']]
```

```python
# updating the values of the dataframe using loc method
df.loc[1, 'Salary'] = 65000
print(df)
```

# iloc

```python

df = pd.DataFrame({
    'Name' : ['Ninja_1','Ninja_2','Ninja_3','Ninja_4'],
    'Department' : ['Marketing','Sales','Marketing','Engineering'],
    'Salary' : [50000,60000,55000,70000]

})
print(df)
```

```python
# Selecting Specific Rows And Columns
# df.iloc[row_positions, column_positions]
row = df.iloc[[1,3], [0,2]]
print(row)
```

```python
# Updating Values

df.iloc[1,2] = 100
print(df)
```

```python
# slicing the dataframe using iloc method
df.iloc[0:3, 0:2]
```
