# Renaming columns 

```python
df = pd.read_csv('student_performance.csv')
df = df.rename(columns={'Math_Score': 'Maths'})
print(df)
```

```python
# you can also rename multiple columns by passing a dictionary of old and new column names
df = df.rename(columns={'Math_Score': 'Maths', 'Science_Score':'Science'})
print(df)
```

```python
# rename index 0, 1 and 2 and update the dataframe
df.set_index('Student_ID',inplace = True)
df.rename(index={'Student_ID':'ID'}, inplace=True)
df
```

# loc and iloc
loc is used for slecting rows or columns based on labels(index) of them
iloc is used selection rows and columns based on integer position of them

```python
df = pd.read_csv('/Users/ypragnesh/Desktop/Pandas_Notes/student_performance.csv')
print(df.head())
```

```python
# we can select particular column from the dataframe using column name
# df.loc[row_labels, column_labels]
print(df.loc[:, 'Math_Score'])
```

```python
df.loc[:, ["Math_Score", "Science_Score"]]
```

# iloc

```python
# df.iloc[row_positions, column_positions]
print(df.iloc[0:5, 0:3])
```
