```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('student_performance.csv')
print(df)
```

```python
# to get the shape of columns
print(df.shape[1])
```

```python
# to get the shape of rows
print(df.shape[0])
```

```python
# checking whether there is null values or not in the DataFrame for that we can use two methods both are same and gives same output
print(df.isnull())
```

```python
print(df.isna())
```

```python
# if we .sum on the null values it will give us the count of null values in each column
print(df.isnull().sum())
```

```python
print(df.isna().sum())
```

```python
# tho drop the columns which have null values we can use dropna() method
df.dropna(inplace=True) # inplace=True will make the changes in the original DataFrame
# if we use df.dropna() without inplace=True it will return a new DataFrame with the dropped columns but the original DataFrame will remain unchanged
print(df)
```

Using inplace=True in pandas is a way to modify the original DataFrame or Series directly without creating a new object in memory. The primary reasons for using it are to update the original variable and, in some rare cases, to potentially save memory when working with very large datasets

```python
# if you want to remove the rows containing null values we can use axis = 0 in the dropna() method
# as we have already removed the rows containing null values in the previous example so there is no change in the DataFrame
# if we didnt specify the axis parameter it will remove the rows containing null values by default
df.dropna(inplace=True,axis = 0)
print(df)
```

```python
df.dropna(inplace=True,axis = 1)
# if you want to remove the columns containing null values we can use axis = 1 in the dropna() method
print(df)
```

```python
# here is another example by using axis = 1 to drop the columns containing null values
df = pd.DataFrame({'A': [1, 2, np.nan], 'B': [4, None, 6]})
print(df.dropna(axis=1))
```

```python
# you can drop null values from specific column by using subset parameter
df = pd.read_csv('student_performance.csv')
df.dropna(subset=['Math_Score'], inplace=True)
print(df)
```

```python
# you can drop null values from specific column by using subset parameter
x = pd.DataFrame({'A': [1, np.nan, 3], 'B': [4, 5, None]})
print(x)
print('\n')
print(x.dropna(subset=['A']))
```

```python
# here is the example with using inplace = True and without using inplace = True
y = pd.DataFrame({'X': [1, np.nan, 3], 'Y': [np.nan, 5, 6]})
y.dropna(inplace=True)
print(y)
```

```python
y = pd.DataFrame({'X': [1, np.nan, 3], 'Y': [np.nan, 5, 6]})
y.dropna()
print(y)
```

```python
# here we are replacing null values with a specific value for that we will use fillna() method
df = pd.read_csv('student_performance.csv')
df.fillna(40,inplace = True)
print(df)
```

```python
df = pd.read_csv('student_performance.csv')
df['Math_Score'] = df['Math_Score'].fillna(df['Math_Score'].mean())
# here we are replacing null values in the Math_Score column with the mean of the Math_Score column
df
```

```python
# this is alternative way to replace null values in the specific column with the mean of that specific column 
# but this is only applicable when you are doing for one column if you want to do for multiple columns then you can use the first method
df.fillna({'Math_Score': df['Math_Score'].mean()}, inplace=True)
```

```python
df['Science_Score'] = df['Science_Score'].fillna(df['Science_Score'].median())
df
```

```python
df['English_Score'] = df['English_Score'].fillna(df['English_Score'].mode())
df
```
