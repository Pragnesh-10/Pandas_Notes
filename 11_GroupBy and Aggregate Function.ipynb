groupby(): Used to group data based on one or more columns before applying an aggregation function to each group

```python
import pandas as pd
```

# agg()

```python
# to calculate the mean of the specific column we can use agg() method and pass the function name as a string to it
# here the function is mean
df['Math_Score'].agg('mean')
```

```python
# to calculate the median of the specific column we can use agg() method and pass the function name as a string to it
# here the function is median
df['Science_Score'].agg('median')
```

```python
# to calculate the minimum value of the specific column we can use agg() method and pass the function name as a string to it
# here the function is min
df['Science_Score'].agg('min')
```

```python
# to calculate the maximum value of the specific column we can use agg() method and pass the function name as a string to it
# here the function is max
df['Science_Score'].agg('max')
```

```python
# to calculate the standard deviation of the specific column we can use agg() method and pass the function name as a string to it
# here the function is std
df['Science_Score'].agg('std')
```

```python
# to claculate number of non null vlaues in the specific column we can use agg() method and pass the function name as a string to it
# here the function is count

df['Science_Score'].agg('count')
```

```python
# to calculate the variance of the specific column we can use agg() method and pass the function name as a string to it
# here the function is var
df['Science_Score'].agg('var')
```

```python
# to calculate the sum of the specific column we can use agg() method and pass the function name as a string to it
# here the function is sum
df['Math_Score'].agg('sum')
```

```python
# here we can pass multiple functions to the agg() method by passing a list of function names as strings to it
df['Math_Score'].agg(['mean','max','min','std','var','count'])
```

```python
# to calculate the aggregate function on multiple columns we can pass a list of column names to the agg() method and then pass the function name as a string to it
df[['Math_Score', 'Science_Score']].agg("sum", axis=1)
# here we are using axis = 1 to calculate sum across the columns for each row
```

```python
#  if we use axis = 0 then it will calculate the sum for each column separately
df[['Math_Score', 'Science_Score']].agg("sum", axis=0)
```

```python
# here we are performing column wise aggregation with multiple functions on multiple columns here default axis = 0
df.agg({'Math_Score' : ['sum', 'mean'], 'Science_Score' : ['sum', 'max']})
# here if you observe the output mean for Science_Score is NaN because we have not specified the function for that column so it will return NaN for that column and for Math_Score we have specified both sum and mean so it will return the sum and mean for that column
# NaN here means “function not applied”, not bad data
```

# Groupby
what groupby does
First, it splits the data into groups.
Second, it applies an aggregation on each group.
Finally, it combines the results into a new table.


split-->Apply-->combine


```python
import pandas as pd

df = pd.DataFrame({
    "gender": ["male", "female", "female", "male", "female"],
    "lunch": ["standard", "free", "standard", "free", "standard"],
    "math_score": [72, 85, 90, 60, 78],
    "reading_score": [70, 88, 92, 65, 80]
})
print(df)
```

```python
# here we are splitting data into groups by using gender column, appling aggregations on math_score column for each group and last we are combining the results into one group
df.groupby("gender")["math_score"].mean()
# here pandas looks at gender column, finds the unique values in the coluymns and splits them into two groups one is male and anoother is female
```

```python
# here we are splitting the data into groups using gender column, we are appling aggregation on mathscore and reading score for each group and we last we are the results into one group
df.groupby("gender")[["math_score", "reading_score"]].mean()
```

```python
# multiple aggregations on multiple columns
df.groupby("gender").agg({
    "math_score": ["mean", "max"],
    "reading_score": "min"
})
```

```python
# here we are splitting data into groups using two columns one is lunch and another is gender and then we are applying aggregation on math_score column for each group and last we are combining the results into one group
df.groupby(["gender", "lunch"])["math_score"].mean()

# interview answering style
# We group data by gender and lunch type, compute the mean math score for each combination, and return a multi-level indexed summary table.”
```

```python
# we can also calculate the size of each group by using size() method
df.groupby("gender").size()
```

```python
# we can also add the the group results to the original dataframe by using transform() method
df["avg_math_by_gender"] = df.groupby("gender")["math_score"].transform("mean")
print(df)
# 
```

```python
# we can also sort the group results by using sort_values() method
df.groupby("gender")["math_score"].mean().sort_values(ascending=True)
df
# we are using ascending = False to sort the values in descending order if we want to sort in ascending order then we can use ascending = True or we can simply omit the ascending parameter because by default it is True
```
